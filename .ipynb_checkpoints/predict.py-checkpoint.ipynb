```python
def drop_low_corr(data):
    corrmat = data.corr().abs().total_price
    high_corr_cols = []
    i = 0
    for value in corrmat:
        if value >= 0.1 :
            high_corr_cols.append(data.columns[i])
        i+=1
   
    
    high_corr_cols.remove('total_price')
    return high_corr_cols

high_corr_cols = drop_low_corr(train_data)
train_data2 = train_data[high_corr_cols]

train_data2 = train_data

print(train_data2.shape)
corrmat = train_data2.corr()
f, ax = plt.subplots(figsize=(40, 9))
sns.heatmap(corrmat, vmax=0.8, square=True)
#sns.heatmap(corrmat, vmax=1, square=True, annot=True, cmap='viridis')
```

# GBDT can also be used as a base model for feature selection in the tree model. (Random Forest)

```python
clf = RandomForestRegressor(n_estimators=100)

def select_feature(X_train, y_train, clf):
    sfm = SelectFromModel(clf)
    sfm.fit(X_train, y_train)
   
        
    return sfm
    
sfm = select_feature(X_train, y_train, clf)

n_X_train = sfm.transform(X_train)
n_X_test = sfm.transform(X_test)
```

```python
n_X_train.shape
```

## Decomposition


```python
from sklearn.decomposition import PCA

def decomposition(n_X_train, n_X_test):
    decomposition_X_train = PCA(n_components=2).fit_transform(n_X_train)
    decomposition_X_test = PCA(n_components=2).fit_transform(n_X_test)
    
    return decomposition_X_train, decomposition_X_test
n_X_train, n_X_test= decomposition(n_X_train, n_X_test)
```

```python
print(n_X_train.shape)
```

## Testing feature selection by RF

```python
clf = RandomForestRegressor(n_estimators=100)

clf.fit(n_X_train, y_train)
y_pred = clf.predict(n_X_test)
```

```python
print('after feature select, prediction value', y_pred[1])
print("after feature select, real value ", y_test[1])

print('Accuracy -->', clf.score(n_X_test, y_test)*100)
print('Actual Loss -->', np.sqrt(mean_squared_error(y_test, y_pred)))

print('data test ', X_test.shape)
print('y_test_pred', y_pred.shape)
```
